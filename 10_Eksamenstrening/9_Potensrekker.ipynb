```python
def factorial(n):
    """This is a recursive function
    to find the factorial of an integer"""

    if n == 0:
        return 1
    else:
        # recursive call to the function
        return n * factorial(n-1)

def taylor_series(x, N):
    
    total = 0
    for n in range(N):
        total += x**n / factorial(n)
    
    return total

for i in range(1, 10):
    print(f"Taylor series with n={i}: {taylor_series(1, i)}")
    
L = 2.71828182845904523536

for i in range(1,100):
    if abs(L - taylor_series(1, i)) < 1e-15:
        print(f"We need a taylor polynomial with n={i} to estimate e")
        break
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.arange(0, 10, 1)

for i in range(1,10):
    plt.plot(x_values, taylor_series(x_values, i), label=f"Taylor {i}")
    
plt.legend()
```

```python
def sin(x, N):
    
    total = 0
    for n in range(N):
        total += ((-1)**(n) * x**(2*n+1)) / factorial(2*n+1)
        
    return total

for n in range(1,10):
    print(f"Sin(1) = {sin(1, n)}")
```
