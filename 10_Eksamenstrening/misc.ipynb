```python
import matplotlib.pyplot as plt

T = 20
N = 1000
h = T / N

x = np.zeros(N+1)
y = np.zeros(N+1)
x[0], y[0] = 1, 1

for i in range(N):
    x_star = x[i] + h * y[i]
    y_star = y[i] - h * np.sin(x[i])
    
    x[i+1] = x[i] + (h/2) * (y[i] + y_star)
    y[i+1] = y[i] - (h/2) * (np.sin(x[i]) + np.sin(x_star))

    
plt.plot(np.linspace(0,T,N+1), x, label="x")
plt.plot(np.linspace(0,T,N+1), y, label="y")
```

```python
T = 200
N = 1000
h = T / N

y = np.zeros(N+1)
y[0] = 0.01

for i in range(N):
    
    y[i+1] = y[i]
    for _ in range(100):
        prev = y[i+1]
        y[i+1] = y[i] + (h/2)*((y[i]**2 - y[i]**3) + (y[i+1]**2 - y[i+1]**3))
        if abs(y[i+1] - prev) < 1e-2:
            break
        
plt.plot(np.linspace(0,T,N+1), y) 
```
