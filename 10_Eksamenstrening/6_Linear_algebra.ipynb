```python
import numpy as np
import matplotlib.pyplot as plt

#data og matrise
x=np.array(range(1,6))
y=np.array([2,4,3,5,6])
z=np.ones(5)
A=np.transpose([x,z])

#regne ut beta_1 og beta_0
beta=np.linalg.solve(np.dot(np.transpose(A),A),np.dot(np.transpose(A),y)) #lÃ¸se system
beta=beta[::-1] #python indekserer jo motsatt av oss, sÃ¥ snu om for lesbarhetens skyld

#regresjonspolynomet
t=np.linspace(0,5,100)
p=beta[1]*t+beta[0]

#plotte
plt.figure()
plt.scatter(x,y) #datapunkter
plt.plot(t,p) #regresjonspolynomet
plt.savefig('reg.png')
```

```python
import numpy as np

A = np.array([[1, 2, 2],
              [2, 6, 2],
              [2, 2, 6]])

B = np.array([[2, 1, 1],
              [1, 2, 1],
              [1, 1, 2]])

# Find A's eigenvalues and eigenvectors
eigvals_A, eigvecs_A = np.linalg.eig(A)
print("A:")
print(f"    {eigvals_A}")
print(f"    {eigvecs_A}")

# Find B's eigenvalues and eigenvectors
eigvals_B, eigvecs_B = np.linalg.eig(B)
print("B:")
print(f"    {eigvals_B}")
print(f"    {eigvecs_B}")
```
